# Hands-on Activity 1

$$ \dfrac{dy}{dt} = ty + t^3, \quad 0 \leq t \leq 1, \quad y(0) = 1. $$

## Euler's Method

$$ y(t_{i+1}) \approx  y(t_i) + h f(t_i,y_i) $$

```python
import numpy as np
import matplotlib.pyplot as plt

def euler_method(f,t0,T,y0,h):
    t = np.arange(t0, T+h, h)
    y = np.zeros(len(t))
    y[0] = y0

    for i in range(len(t)-1):
        y[i+1] = y[i] + h*f(t[i],y[i])

    return t, y
```

```python
def f(t, y):
  return t*y+t**3

t0 = 0
T = 1
y0 = 1
h = 0.1

t, y = euler_method(f, t0, T, y0, h)

# Plot the solution
plt.plot(t, y)
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Solution of dy/dt = ty+t^3 using Euler's method")
plt.grid(True)
plt.show()
```

## Runge-Kutta Method of Order 4 (RK4)

$$ y_{i+1} = y_i + \dfrac{h}{6}(s_1 + 2s_2 + 2s_3 + s_4), $$
where
$$ \begin{array}{rcl}
s_1 & = & f(t_i,y_i) \\
s_2 & = & f \left ( t_i + \dfrac{h}{2}, y_i + \dfrac{h}{2} s_1 \right ) \\
s_3 & = & f \left ( t_i + \dfrac{h}{2}, y_i + \dfrac{h}{2} s_2 \right ) \\
s_4 & = & f( t_{i+1}, y_i + h s_3 )
\end{array} . $$

```python
def RK4_method(f,t0,T,y0,h):
    t = np.arange(t0, T+h, h)
    y = np.zeros(len(t))
    y[0] = y0

    for i in range(len(t)-1):
        s1 = f(t[i],y[i])
        s2 = f(t[i] + h/2, y[i] + s1*h/2)
        s3 = f(t[i] + h/2, y[i] + s2*h/2)
        s4 = f(t[i+1], y[i] + s3*h)
        y[i+1] = y[i] + (h/6)*(s1 + 2*s2 + 2*s3 + s4)

    return t, y
```

$$ \dfrac{dy}{dt} = ty + t^3, \quad 0 \leq t \leq 1, \quad y(0) = 1. $$

```python
def f(t, y):
  return t*y+t**3

t0 = 0
T = 1
y0 = 1
h = 0.1

t, y = RK4_method(f, t0, T, y0, h)

# Plot the solution
plt.plot(t, y)
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Solution of dy/dt = ty+t^3 using RK4 method")
plt.grid(True)
plt.show()
```

## Implicit Euler's method

$$ y(t_{i+1}) \approx  y(t_i) + h f(t_{i+1},y_{i+1}) $$

Use Implicit Euler's method, with $h = 0.1$, to solve the IVP
$$ \dfrac{dy}{dt} = ty + t^3, \quad 0 \leq t \leq 1, \quad y(0) = 1. $$
which results to

$$ y_{i+1} = \dfrac{y_i + ht_{i+1}^3}{1 - ht_{i+1}}$$

```python
t0 = 0
T = 1
y0 = 1
h = 0.1

t = np.arange(t0, T+h, h)
y = np.zeros(len(t))
y[0] = y0

for i in range(len(t)-1):
    y[i+1] = (y[i]+h*t[i+1]**3)/(1-h*t[i+1])

# Plot the solution
plt.plot(t, y)
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Solution of dy/dt = ty+t^3 using Implicit Euler method")
plt.grid(True)
plt.show()
```
