3. Solve the two-point BVP using the finite difference method with $h = 0.2$
$$ y'' = 4y, \ 0 \leq t \leq 1, $$
with boundary conditions
$$ y(0) = 1, \qquad y(1) = 3. $$

```python
import numpy as np
from numpy.linalg import solve
import matplotlib.pyplot as plt

N = 4
h = 1/(N+1)

t0 = 0
T = 1

t=np.arange(t0,T+h,h)
d1 = (-2-4*(h**2))*np.ones(N) #main diagonal (-2-4h^2 is the main diagonal)
d2 = np.ones(N-1)#above and below diagonal
A = np.diag(d1) + np.diag(d2,1) + np.diag(d2,-1) #1 if above main diagonal and -1 for below main diagonal
b = np.zeros(N) #starts at 1 (since we know t0) and N+1 ends
b[-1] = -3
b[0] = -1

print(A)
print(b)

y = solve(A,b)
print(y)
plt.figure(1)
plt.plot(t[1:N+1],y)
y = np.pad(y, 1)
y[-1] = 3
y[0]= 1
print(y)

# Plot the solution
plt.figure(2)
plt.plot(t, y)
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Solution of y'' = 4y using finite difference method")
plt.grid(True)
plt.show()
```
