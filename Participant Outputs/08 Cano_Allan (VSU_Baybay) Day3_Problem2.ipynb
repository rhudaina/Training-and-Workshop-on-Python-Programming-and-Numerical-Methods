Convert $y'' - ty = 0$ to an equivalent system of IVP and solve it on the interval  $[0,1]$ using Euler's method with $h = 0.25$ and initial conditions $y(0) = y'(0) = 1$.
   

```python
def RK4_method(F,t0,T,Y0,h):
    t = np.arange(t0, T+h, h)
    Y = np.zeros((len(Y0),len(t)))
    Y[:,0] = Y0

    for i in range(len(t)-1):
        s1 = F(t[i],Y[:,i])
        s2 = F(t[i] + h/2, Y[:,i] + s1*h/2) #middle point
        s3 = F(t[i] + h/2, Y[:,i] + s2*h/2)
        s4 = F(t[i+1], Y[:,i] + s3*h)
        Y[:,i+1] = Y[:,i] + (h/6)*(s1 + 2*s2 + 2*s3 + s4)

    return t, Y
```

```python
def F(t, Y):
    f1 = Y[1]
    f2 = t*Y[0]
    return np.array( [f1,f2])

t0 = 0
T = 1
Y0 = np.array([1,1])
h = 0.25

t, Y = RK4_method(F,t0,T,Y0,h)

print(Y)

# Plot the solution of the system
plt.figure(1)
plt.plot(t, Y[0,:], label='$y_1(t)$')
plt.plot(t, Y[1,:], label='$y_2(t)$')
plt.legend()


# Plot the solution in the context of higher-order DEs
plt.figure(2)
plt.plot(t, Y[0,:])
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Solution of y'' = ty using RK4 method")
plt.grid(True)
plt.show()
```
