# Differential Equations of order one

## Euler's Method

$$ y(t_{i+1}) \approx  y(t_i) + h f(t_i,y_i) $$

```python
import numpy as np
import matplotlib.pyplot as plt

def euler_method(f,t0,T,y0,h):
    t = np.arange(t0, T+h, h)
    y = np.zeros(len(t))
    y[0] = y0

    for i in range(len(t)-1):
        y[i+1] = y[i] + h*f(t[i],y[i])

    return t, y
```

Example: Use Euler's method, with h = 0.1, to solve the IVP

$$ \dfrac{dy}{dt} = -y, \quad 0 \leq t \leq 5, \quad y(0) = 1. $$

```python
def f(t, y):
  return -y

t0 = 0
T = 5
y0 = 1
h = 0.1

t, y = euler_method(f, t0, T, y0, h)

# Plot the solution
plt.plot(t, y)
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Solution of dy/dt = -y using Euler's method")
plt.grid(True)
plt.show()
```

# Runge-Kutta Method of Order 4 (RK4)

$$ y_{i+1} = y_i + \dfrac{h}{6}(s_1 + 2s_2 + 2s_3 + s_4), $$
where
$$ \begin{array}{rcl}
s_1 & = & f(t_i,y_i) \\
s_2 & = & f \left ( t_i + \dfrac{h}{2}, y_i + \dfrac{h}{2} s_1 \right ) \\
s_3 & = & f \left ( t_i + \dfrac{h}{2}, y_i + \dfrac{h}{2} s_2 \right ) \\
s_4 & = & f( t_{i+1}, y_i + h s_3 )
\end{array} . $$

```python
def RK4_method(f,t0,T,y0,h):
    t = np.arange(t0, T+h, h)
    y = np.zeros(len(t))
    y[0] = y0

    for i in range(len(t)-1):
        s1 = f(t[i],y[i])
        s2 = f(t[i] + h/2, y[i] + s1*h/2)
        s3 = f(t[i] + h/2, y[i] + s2*h/2)
        s4 = f(t[i+1], y[i] + s3*h)
        y[i+1] = y[i] + (h/6)*(s1 + 2*s2 + 2*s3 + s4)

    return t, y
```

Example: Use RK4 method, with h = 0.1, to solve the IVP

$$ \dfrac{dy}{dt} = -y, \quad 0 \leq t \leq 5, \quad y(0) = 1. $$

```python
def f(t, y):
  return -y

t0 = 0
T = 5
y0 = 1
h = 0.1

t, y = RK4_method(f, t0, T, y0, h)

# Plot the solution
plt.plot(t, y)
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Solution of dy/dt = -y using RK4 method")
plt.grid(True)
plt.show()
```

## Implicit Euler's method

$$ y(t_{i+1}) \approx  y(t_i) + h f(t_{i+1},y_{i+1}) $$

Example: Use Implicit Euler's method, with h = 0.1, to solve the IVP
$$ \dfrac{dy}{dt} = -y, \quad 0 \leq t \leq 5, \quad y(0) = 1. $$

```python
t0 = 0
T = 5
y0 = 1
h = 0.1

t = np.arange(t0, T+h, h)
y = np.zeros(len(t))
y[0] = y0

for i in range(len(t)-1):
    y[i+1] = y[i]/(1+h)

# Plot the solution
plt.plot(t, y)
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Solution of dy/dt = -y using Implicit Euler method")
plt.grid(True)
plt.show()
```

# Higher Order Differential Equations 

## Euler's method

$$ \left \{ \begin{array}{rcl}
y^{i+1}_{1} & = & y^{i}_{1} + h f_1(t,y^{i}_{1},y^{i}_{2},\ldots,y^{i}_{m}), \\
y^{i+1}_{2} & = & y^{i}_{2} + h f_2(t,y^{i}_{1},y^{i}_{2},\ldots,y^{i}_{m}), \\
 & \vdots & \\
y^{i+1}_{m} & = & y^{i}_{m} + h f_m(t,y^{i}_{1},y^{i}_{2},\ldots,y^{i}_{m}),
\end{array} \right . $$

$$
Y^{i+1} = Y^i + h*F(t,Y)
$$

Example: Convert $y'' - ty = 0$ to an equivalent system of IVP and solve it on the interval  $[0,1]$ using Euler's method with $h = 0.25$ and initial conditions $y(0) = y'(0) = 1$.

Solution: 

```python
def euler_sys(F,t0,T,Y0,h):
    t = np.arange(t0, T+h, h)
    Y = np.zeros((len(Y0),len(t)))
    Y[:,0] = Y0

    for i in range(len(t)-1):
        Y[:,i+1] = Y[:,i] + h*F(t[i],Y[:,i])

    return t, Y
```

```python
def F(t, Y):
    f1 = Y[1]
    f2 = t*Y[0]
    return np.array( [f1,f2])

t0 = 0
T = 1
Y0 = np.array([1,1])
h = 0.25

t, Y = euler_sys(F,t0,T,Y0,h)

print(Y)

# Plot the solution of the system
plt.figure(1)
plt.plot(t, Y[0,:], label='y1(t)')
plt.plot(t, Y[1,:], label='y2(t)')


# Plot the solution in the context of higher-order DEs
plt.figure(2)
plt.plot(t, Y[0,:])
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Solution of y'' = ty using Euler method")
plt.grid(True)
plt.show()
```
